
# Background: 
Breast cancer is traditionally stratified using immunohistochemical subtypes.  

# Research question: 
Can different subtype classification information from the METABRIC dataset improve survival prediction beyond conventional immunohistological classification?

# Hypothesis:
Subtype information improves models predicting patient survival, compared to immunohistochemical and clinical data.  

# Approach
1. Create a model to predict overall survival (months) from baseline data and immunohistological data (control)
2. Create a model to predict overall survival (months) from baseline and 3-gene subtype 
3. Create a model to predict overall survival (months) from baseline and PAM50 + claudin data
4. Create a model to predict overall survival (months) from baseline and Integrative Cluster data
5. Compare these models to see which is best
